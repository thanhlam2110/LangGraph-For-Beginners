```python
from typing import TypedDict
from langgraph.graph import END, START, StateGraph
from IPython.display import Image, display


# 1. Định nghĩa kiểu state (trạng thái)
# Định nghĩa một state (trạng thái) mà các node trong graph sẽ nhận và trả về.
# Ở đây state chỉ đơn giản là một dict có một trường "message" kiểu str.
class HelloWorldState(TypedDict):
    message: str
    id: int


# 2. Định nghĩa các node (tác vụ) của graph
# hello và bye là hai hàm đại diện cho hai node (nút) trong graph.
# Mỗi node nhận vào một state (dict), in ra nội dung, rồi trả về state mới với chuỗi "Hello ..." hoặc "Bye ...".
# Logic đơn giản để minh hoạ cách truyền state qua từng bước (node).

def hello(state: HelloWorldState):
    print(f"Hello Node: {state['message']}-{state['id']}")
    return {"message": "Hello "+state['message']+"-"+str(state['id'])}


def bye(state: HelloWorldState):
    print(f"Bye Node: {state['message']}")
    return {"message": "Bye "+state['message']}

# 3. Xây dựng workflow (đồ thị)
# Tạo một StateGraph (tức là workflow dạng đồ thị).
# Thêm 2 node vào đồ thị: "hello" và "bye".
# Nối các node theo thứ tự: START → hello → bye → END.
# Luồng xử lý sẽ đi từ điểm bắt đầu (START), vào node "hello", rồi node "bye", rồi kết thúc (END).

graph = StateGraph(HelloWorldState)
graph.add_node("hello",hello)
graph.add_node("bye",bye)

graph.add_edge(START,"hello")
graph.add_edge("hello","bye")
graph.add_edge("bye",END)

# 4. Biên dịch graph thành runnable workflow
runnable = graph.compile()

# 5a. (Tuỳ chọn) Lưu hình ảnh workflow ra file PNG
## Save graph to file

# img_bytes = runnable.get_graph().draw_png()
# with open("simple-workflow.png", "wb") as f:
#     f.write(img_bytes)
# print("Save simple-workflow.png")

# 5b. Hiển thị workflow trực tiếp trong notebook
## Display graph
try:
    display(Image(runnable.get_graph().draw_png()))
except Exception:
    pass
# 6. Chạy workflow với dữ liệu thực
# Chạy workflow với input ban đầu là {"message": "Lam"}.
# Luồng đi như sau:
# Node "hello" nhận {"message": "Lam"} → trả về {"message": "Hello Lam"}
# Node "bye" nhận {"message": "Hello Lam"} → trả về {"message": "Bye Hello Lam"}
# Kết quả in ra: {'message': 'Bye Hello Lam'}

output = runnable.invoke({"message": "Lam","id":123})
print(output)
```
